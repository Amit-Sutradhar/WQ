```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file into a DataFrame
file_path = 'STP PH2.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Data Visualization
# Line Plot for pH Probe 1
plt.figure(figsize=(12, 6))
sns.lineplot(x='TIME', y='pH Probe 1', data=df, label='pH Probe 1')
sns.lineplot(x='TIME', y='PH ', data=df, label='PH ')
plt.title('pH Probe 1 vs PH  over Time')
plt.xlabel('TIME')
plt.ylabel('pH Value')
plt.legend()
plt.show()

# Line Plot for pH Probe 2
plt.figure(figsize=(12, 6))
sns.lineplot(x='TIME', y='pH Probe 2', data=df, label='pH Probe 2')
sns.lineplot(x='TIME', y='PH ', data=df, label='PH ')
plt.title('pH Probe 2 vs PH  over Time')
plt.xlabel('TIME')
plt.ylabel('pH Value')
plt.legend()
plt.show()

# Line Plot for pH Probe 3
plt.figure(figsize=(12, 6))
sns.lineplot(x='TIME', y='pH Probe 3', data=df, label='pH Probe 3')
sns.lineplot(x='TIME', y='PH ', data=df, label='PH ')
plt.title('pH Probe 3 vs PH  over Time')
plt.xlabel('TIME')
plt.ylabel('pH Value')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file into a DataFrame
file_path = 'STP PH2.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Data Visualization
plt.figure(figsize=(12, 6))

# Line Plot for all pH probes
sns.lineplot(x='TIME', y='pH Probe 1', data=df, label='pH Probe 1', color='blue')
sns.lineplot(x='TIME', y='pH Probe 2', data=df, label='pH Probe 2', color='orange')
sns.lineplot(x='TIME', y='pH Probe 3', data=df, label='pH Probe 3', color='green')
sns.lineplot(x='TIME', y='PH ', data=df, label='PH ', color='red')

plt.title('pH Probes vs PH  over Time')
plt.xlabel('TIME')
plt.ylabel('pH Value')
plt.legend()
plt.show()
```
